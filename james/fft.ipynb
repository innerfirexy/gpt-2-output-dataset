```python
from scipy.fft import fft, fftfreq, fftshift
from scipy import signal

import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(256)
print(t)
```

```python
plt.plot(t, np.sin(t))
plt.show()
```

```python
freq_unshifted = fftfreq(t.shape[-1])
print(freq_unshifted.shape)
freq_unshifted
```

```python
freq = fftshift(fftfreq(t.shape[-1]))
print(freq.shape)
freq
```

```python
sp = fftshift(fft(np.sin(t)))
plt.plot(freq, sp.real)
```

```python
### https://blog.csdn.net/s09094031/article/details/105744720


input_file = '/Users/james/Workspace/gpt-2-output-dataset/data/webtext.test.model=.nll'
with open(input_file, 'r') as f:
    line = f.readlines()
    line = line[1642] # short 9
x = np.array(list(map(float, line.strip().split())))
print(x.shape)


freq_x = fftshift(fftfreq(x.shape[-1]))
sp_x = fftshift(fft(x))
print('freq_x:', freq_x.shape)
print('sp_x:', sp_x.shape)
plt.plot(freq_x[len(freq_x)//2:], sp_x[len(sp_x)//2:], label='x')

y = np.pad(x, (0, 100), 'constant')
freq_y = fftshift(fftfreq(y.shape[-1]))
sp_y = fftshift(fft(y))
print('freq_y:', freq_y.shape)
print('sp_y:', sp_y.shape)
plt.plot(freq_y[len(freq_y)//2:], sp_y[len(sp_y)//2:], label='y')

from scipy import interpolate
f = interpolate.interp1d(freq_x[len(freq_x)//2:].tolist(), sp_x[len(sp_x)//2:].tolist(), fill_value="extrapolate")
freq_z = np.linspace(0, 0.5, 342)
sp_z = f(freq_z)

print('freq_z:', freq_z.shape)
print('sp_z:', sp_z.shape)
plt.plot(freq_z, sp_z, label='z')


plt.legend()
```
